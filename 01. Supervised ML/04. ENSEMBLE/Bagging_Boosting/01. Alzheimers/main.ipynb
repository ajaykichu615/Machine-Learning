Sampling

```python
X=df.drop(columns='alzheimers_diagnosed')
y=df['alzheimers_diagnosed']
```

```python
pipe=Pipeline([
    ('oversampling', SMOTE(sampling_strategy={1:800})),
    ('undersampling', RandomUnderSampler(sampling_strategy={0:800}))
])
```

```python
X_sampled, y_sampled=pipe.fit_resample(X,y)
```

```python
y_sampled.value_counts()
```

Scaling

```python
minmax=MinMaxScaler()
X_scaled=minmax.fit_transform(X_sampled)
```

Splitting

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled, y_sampled, test_size=0.3, random_state=42)
```

Model Selection

```python
models={KNeighborsClassifier(), SVC(), DecisionTreeClassifier(), GaussianNB(), RandomForestClassifier(), GradientBoostingClassifier()}
for model in models:
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    print(model)
    print(round(accuracy_score(y_test, y_pred)*100))
    print(classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
```

Cross Validation

Cross validation is used to evaluate the performance and generalization ability of a machine learning model. It gives more reliable estimate of model performace by using multiple train-test-split instead of just one.

```python
from sklearn.model_selection import cross_val_score
print((cross_val_score(GaussianNB(), X_scaled, y_sampled,cv=5).mean())*100) # Mean accuracy score of 5 gaussian models
```
