```python
data = {
    'CustomerID': range(101, 121),
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ian', 'Jane',
             'Kyle', 'Laura', 'Mark', 'Nina', 'Oscar', 'Paula', 'Quinn', 'Rachel', 'Steve', 'Tina'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 
               'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Chennai', 'Delhi', 'Mumbai',
             'Bangalore', 'Delhi', 'Mumbai', 'Chennai', 'Delhi', 'Chennai', 'Bangalore', 'Delhi', 'Mumbai', 'Chennai'],
    'PurchaseAmount': [1200, 540, 760, 1340, 2300, 890, 500, 650, 970, 1430,
                       3100, 780, 990, 660, 400, 2000, 1800, 750, 1450, 920],
    'LoyaltyPoints': [120, 50, 76, 130, 230, 89, 50, 65, 97, 143,
                      310, 78, 99, 66, 40, 200, 180, 75, 145, 92]
}

df = pd.DataFrame(data)
```

```python
df
```

🔍 Filtering with .loc[]

1. Filter customers who are from Mumbai.


```python

customers_mumbai = df.loc[df['City'] == 'Mumbai']
customers_mumbai
```

2. Show all rows where PurchaseAmount is greater than 1000.

```python
Purchase_amt = df.loc[df['PurchaseAmount']>1000]
Purchase_amt
```

3. Filter female customers who are from Delhi.

```python
female_customers = df.loc[(df['Gender'] == 'F') & (df['City']=='Delhi')]
female_customers
```

4. Display names and purchase amounts of customers who spent less than ₹800.

```python
low_spenders = df.loc[df['PurchaseAmount'] < 800, ['Name', 'PurchaseAmount']]
low_spenders
```

5. Show all data for customers who have LoyaltyPoints >= 100 and are from Bangalore.

```python
loyalty_points_100 = df.loc[(df['LoyaltyPoints']>=100) & (df['City']=='Bangalore')]
loyalty_points_100
```

6. List customers whose names start with 'A' or 'M'.

```python
customers_AM = df.loc[df['Name'].str.startswith(('A', 'M'))]
customers_AM
```

7. Find male customers who spent between ₹500 and ₹1000.

```python
male_cust = df.loc[(df['Gender'] == 'M') & (df['PurchaseAmount'] > 500) & (df['PurchaseAmount'] < 1000)]
male_cust 
```

8. Filter customers not from Chennai.

```python
customers_not_chennai = df.loc[df['City'] != 'Chennai'][['Name']]
customers_not_chennai
```

9. Show customers who are from Delhi or Mumbai and have LoyaltyPoints above 90.

```python
customers_DM = df.loc[((df['City'] == 'Delhi') | (df['City'] == 'Mumbai')) & (df['LoyaltyPoints'] > 90)]
customers_DM
```

10. Filter customers with even-numbered CustomerID.

```python
customerid_even = df.loc[df['CustomerID']%2==0]
customerid_even
```

✏ Manipulation using .loc[]

11. Set LoyaltyPoints to 0 for all customers from Chennai.

```python
df.loc[df['City'] == 'Chennai', 'LoyaltyPoints'] = 0
df.loc[df['City']=='Chennai']
```

12. Add ₹200 to PurchaseAmount for all female customers.

```python
df.loc[df['Gender'] == 'F', 'PurchaseAmount'] += 200
df.loc[df['Gender']=='F']
```

13. Update the city to 'Hyderabad' for customer CustomerID = 105.

```python
df.loc[df['CustomerID'] == 105, 'City'] = 'Hyderabad'
df.loc[df['CustomerID'] == 105]
```

14. Add 10 bonus points to LoyaltyPoints for customers who spent more than ₹2000.

```python
df.loc[df['PurchaseAmount']>2000, 'LoyaltyPoints'] += 10
df.loc[df['PurchaseAmount']>2000]
```

15. Set name to 'Anonymous' for all customers who spent less than ₹600.

```python
df.loc[df['PurchaseAmount']<600, 'Name'] = 'Anonymous'
df.loc[df['PurchaseAmount']<600]
```

16. Change Gender from 'F' to 'Female' and 'M' to 'Male' using .loc[].

```python
df.loc[df['Gender'] == 'F', 'Gender'] = 'Female' # Another method: df['Gender']=df['Gender'].apply(func=lambda
                                                 #  x:'Female' if x=='F' else 'Male')             
df.loc[df['Gender'] == 'M', 'Gender'] = 'Male'
df
```

17. For customers from Delhi, increase their PurchaseAmount by 10%.

```python
df.loc[df['City']=='Delhi', 'PurchaseAmount'] *= 1.1
df.loc[df['City']=='Delhi']
```

18. Change the city to 'Tier-1' for all customers from Delhi, Mumbai, or Bangalore.

```python
df.loc[df['City'].isin(['Delhi', 'Mumbai', 'Bangalore']), 'City'] = 'Tier-1'
df['City']
```

19. Set LoyaltyPoints to 'High' if above 100, else 'Low'.

```python
df.loc[df['LoyaltyPoints']>100, 'LoyaltyPoints'] = 'High'
df['LoyaltyPoints']
```

20. Add a new column 'RewardEligible' where value is 'Yes' if PurchaseAmount > 1000 else 'No'.

```python
df.loc[df['PurchaseAmount'] > 1000, 'RewardEligible'] = 'Yes'
df.loc[df['PurchaseAmount'] <= 1000, 'RewardEligible'] = 'No'
df
```
