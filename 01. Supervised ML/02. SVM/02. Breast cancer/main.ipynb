*Decided not to remove outliers since it might affect the overall dataset significantly.*

Label Encoding

```python
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

Scaling

```python
X=df.drop(columns='diagnosis')
y=df['diagnosis']
```

```python
minmax=MinMaxScaler()
X_scaled=minmax.fit_transform(X)
```

Oversampling

```python

smote=SMOTE()
X_over,y_over=smote.fit_resample(X_scaled,y) # type: ignore
```

```python
pd.DataFrame(X_over).duplicated().sum() # to verify if the SMOTEd values are duplicates or not
```

```python
y_over.value_counts()
```

Data Splitting

```python
X_train, X_test, y_train, y_test= train_test_split(X_scaled,y,test_size=0.3, random_state=42)
```

Model Training

```python
svm_model = SVC()
```

```python
svm_model.fit(X_train, y_train)
```

Prediction

```python
y_pred_svm = svm_model.predict(X_test)
```

```python
print(round(accuracy_score(y_test, y_pred_svm)*100,2),"%",sep=' ')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_svm,zero_division=0))
```
