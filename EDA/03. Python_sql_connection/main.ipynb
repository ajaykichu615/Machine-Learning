```python
import pymysql
```

```python
connection=pymysql.connect(host='localhost', user='root', 
                           password='2255', database='walmart')
```

```python
cursor=connection.cursor()
```

```python
cursor.execute("select id, city, gender, total from walmart_sales")
```

~ to fetch data

```python
data=cursor.fetchall()
```

```python
print(data)
```

##### Now the data can be manipulated further using Python
~ data from sql is fetched as tuples, so below we parse it further by manipulating tuples

```python
for a,b,c,d in data:
    if c=='Male':
        print(a)
```

~ to create a table

```python
cursor.execute('create table walmart_b (id int primary key auto_increment, name_of_product varchar(50), price int, quantity int)')
```

~ Unless I commit the query, the data won't be inserted into the table - even though the
query shows successful

```python
cursor.execute('Insert into walmart_b(name_of_product, price, quantity) values(%s,%s,%s)',('Bag',5000, 4))
connection.commit()
```
